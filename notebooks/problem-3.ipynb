Let $G_1$ have vertex set $V(G_1) = \{A,B,C,D\}$, where $\{A,B\}$ and $\{C,D\}$ are independent sets (i.e. no two vertices in the subset are adjacent). Let $G_2$ have vertex set $V(G_2) = \{0,1,2,3\}$ such that the set forms a cycle.

Based on how we've defined the vertex sets of $G_1$ and $G_2$, we define the edge sets as:
- $E(G_1) = \{(A,C),(A,D),(B,C),(B,D)\}$,
- $E(G_2) = \{(0,1),(1,2),(2,3),(3,0)\}$. 

We can now construct a bijection $f: V(G_1) \to V(G_2)$
- $f(A) = 0$, 
- $f(B) = 2$, 
- $f(C) = 1$, 
- $f(D) = 3$.

This gives us a valid edge mapping: 
- $(A,C)$ maps to $(0,1)$, 
- $(A,D)$ maps to $(0,3)$, 
- $(B,C)$ maps to $(2,1)$, 
- $(B,D)$ maps to $(2,3)$. 

Therefore, $G_1$ is isomorphic to $G_2$
